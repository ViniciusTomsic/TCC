# Verificação dos Datasets Gerados

Este notebook carrega os datasets de treino e teste gerados e verifica a distribuição das classes.

```python
import pandas as pd
import os

# Define paths - assumindo que o notebook está na raiz ou src, vamos buscar data/ relativo ao cwd
base_path = os.getcwd()
data_dir = os.path.join(base_path, 'data')
train_path = os.path.join(data_dir, 'df_treino.pkl')
test_path = os.path.join(data_dir, 'df_teste.pkl')

print(f"Procurando dados em: {data_dir}")
```

```python
# Carregar DataFrames
if os.path.exists(train_path) and os.path.exists(test_path):
    df_treino = pd.read_pickle(train_path)
    df_teste = pd.read_pickle(test_path)

    print(f"Shape df_treino: {df_treino.shape}")
    print(f"Shape df_teste: {df_teste.shape}")
else:
    print("Arquivos não encontrados! Verifique se rodou a geração de dados e se o caminho está correto.")
```

```python
# Verificar Distribuição de Classes no Treino
if 'df_treino' in locals():
    print("Distribuição de classes em df_treino:")
    distrib = df_treino['tipo_falha_adicionada'].value_counts()
    print(distrib)
    
    print("\nPorcentagem:")
    print(df_treino['tipo_falha_adicionada'].value_counts(normalize=True) * 100)
```

```python
# Verificar Distribuição de Classes no Teste
if 'df_teste' in locals():
    print("Distribuição de classes em df_teste:")
    print(df_teste['tipo_falha_adicionada'].value_counts())
```
